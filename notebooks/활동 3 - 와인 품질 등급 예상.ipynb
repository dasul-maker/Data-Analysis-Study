```python
from scipy import stats
from statsmodels.formula.api import ols, glm
#함수 로드
red_wine_quality = wine.loc[wine['type'] == 'red', 'quality']
white_wine_quality = wine.loc[wine['type'] == 'white', 'quality']
#레드/화이트 와인 샘플의 quality 값만 찾아서 각각 red_wine/white_wine에 저장
stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var = False)
#scipy 패키지의 stats.ttest_ind() 함수를 사용하여 t-검정을 하고 두 그룹 간의 차이를 확인
#p-value가 0.05보다 작으면 두 그룹 간의 차이가 있는 것으로 확인
Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol'
#종속 변수 = quality, 독립 변수 = type와 quality를 제외한 11개 속성
#종속 변수 ~~ 독립 변수1 + 독립 변수2 ... + 독립 변수n
regression_result = ols(Rformula, data = wine).fit()
#선형 회귀 모델 중 ols 모델 사용
#statsmodels.formula.api 패키지의 ols() 함수는 첫 번째 인자로 종속 변수와 독립 변수를 구성한 변수를 줌
#두 번째 인자는 실제 사용할 변수값을 가진 데이터프레임 지정
#완성된 선형 회귀 모델은 fit() 함수에 의해 실행되어 regression_result에 저장
regression_result.summary()
#통계값 출력

# Dep.Variable : 종족변수
# Model : 회귀 모델 종류
# Method : 회귀 분석 함수
# Time : 회귀 분석 수행 시간
# No. Observations : 샘플 개수 : 총 표본 수
# Df Residuals : 잔차의 자유도 : 총 표본 수 - 종족 변수 개수 - 독립 변수 개수
# Df Model : 모델의 자유도 : 독립 변수 개수
```

```python
#회귀 분석 모델로 새로운 샘플의 품질 등급 예측
sample1 = wine[wine.columns.difference(['quality', 'type'])]
#wine에서 quality와 type 열은 제외하고 독립 변수만 추출하여 sample1에 저장
sample1 = sample1[0:5][:]
#sample1에 있는 샘플 중 5개 (0번~4번) 샘플만 추출, sample1에 다시 저장하여 예측에 사용할 샘플 제작
sample1_predict = regression_result.predict(sample1)
#샘플 데이터를 regression_result의 예측 함수 predict()에 적용하여 수행한 뒤 결과 예측값을 sample1_predict에 저장
sample1_predict
```

```python
sample2_predict = regression_result.predict(sample2)
#샘플 데이터를 regression)result의 예측 함수 predict()에 적용하여 수행한 뒤 결과 예측값을 sample2_predict에 저장
sample2_predict
```

```python
#결과 시각화
#와인 유형에 따른 품질 등급 히스토그램 그리기
import matplotlib.pyplot as plt
import seaborn as sns
#패키지 로드
sns.set_style('dark')
#히스토그램 차트의 배경색 스타일 지정
sns.histplot(red_wine_quality, stat = 'density', kde = True, color = 'red', label = 'red wine')
sns.histplot(white_wine_quality, stat = 'density', kde = True, label = 'white wine')
#레드/화이트 와인에 대한 histplot 객체 생성
plt.title("quality of Wine Type")
#차트 제목 설정
plt.legend()
#차트 범례 설정
plt.show()
```

```python
#부분 회귀 플롯으로 시각화
import statsmodels.api as sm

others = list(set(wine.columns).difference(set(["quality", "fixed_acidity"])))
#부분 회귀에 사용할 독립 변수 fixed_acidity와 종속 변수를 제외한 나머지 변수 이름을 리스트 others로 추출
p, resids = sm.graphics.plot_partregress("quality", "fixed_acidity", others, data = wine, ret_coords = True)
#나머지 변수는 고정하고 fixed_acidity가 종속 변수 quality에 미치는 영향에 부분 회귀를 수행
plt.show()
```

```python
fig = plt.figure(figsize = (8, 13))
#차트의 크기 지정
sm.graphics.plot_partregress_grid(regression_result, fig = fig)
#다중 선형 회귀 분석 결과를 가지고 있는 regression_result를 이용해 각 독립 변수의 부분 회귀 플롯을 구함
plt.show()
```
